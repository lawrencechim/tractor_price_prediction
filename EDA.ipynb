```python
mode_enclosure = df['Enclosure'].mode()
df['Enclosure'].fillna(mode_enclosure.values[0], inplace = True)
df.loc[df['Enclosure'] == 'None or Unspecified', 'Enclosure'] = mode_enclosure.values[0]
df['Enclosure'].unique()
enclosures = pd.get_dummies(df['Enclosure'])
enclosures['EROPS AC'] +=  enclosures['EROPS w AC']
enclosures.drop('EROPS w AC', inplace=True, axis=1)
enclosures
df[enclosures.columns[:-1]] = enclosures[enclosures.columns[:-1]]
df
```
